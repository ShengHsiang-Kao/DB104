# User_based_協同過濾推薦函數

```python
def CF_Recommend_User_Based(my_card):
    #imput data type:my_card=['上海商銀-小小兵Bello回饋卡','中國信託-ANA聯名卡']
    hold_card_df=pd.DataFrame(columns=df.columns)
    hold_card_df.loc[0]=0
    #搜尋使用者持有卡於資料庫並設成1
    hold_card_df.loc[0,my_card]=1
    #計算cosine相似度
    c=cosine_similarity(hold_card_df,df)
    # df_m2m = pd.DataFrame(cosine_similarity(df))
    card_array = list(c[0])
    #找出最相似的一筆，印出該筆資料中使用者未持有之卡片(值為0之卡片)
    # print(temp_c)
    max_sim_value=(max(card_array))
    max_sim_user=card_array.index(max_sim_value)
    # c_sort=sorted(temp_c,reverse=True)
    recommend_card=[]
    #找最相似人的持有卡
    for name_index, i in enumerate(df.loc[max_sim_user]):
        #如果最相似的人的卡在使用者未出現
        card=hold_card_df.columns[name_index]
        if i >0 and hold_card_df.loc[0,card] !=1:
            recommend_card.append(card)
    if recommend_card==[]:
        return '沒有好推薦的卡'
    else:
        return recommend_card
```
