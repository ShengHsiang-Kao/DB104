```python
#每個新聞(list)到stop wold比對
ngnews01=[]
ngnewsW01=[]
try:
    for n,i in enumerate(df['list']):
        ngnews=[]     #為了每篇只記錄一次
        for word in i:
            if word in stop:
                if n not in ngnews:
                    ngnews.append(n)
                    ngnews01.append(n)    #號碼
                    ngnewsW01.append(word)  #字體
except:
    pass
```
