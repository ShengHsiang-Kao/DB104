```python
import requests
from bs4 import BeautifulSoup
import json
import os
import re
import pandas as pd
```

```python
def rename(title):
    banned_list = [':','"','<','>','?','\\','/','*','|']
    for iii in banned_list:
        if iii in title:
            title =  title.replace(iii,'')
    return title
```

```python
# cardpath=r'./cardU_news'
# if not os.path.exists(cardpath):
#     os.mkdir(cardpath)

useragent= 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_13_6) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/77.0.3865.90 Safari/537.36'
headers={'User-Agent':useragent}


index = 0
pk= [4,8,58,7,26,6,5,27,28]

x=1
df= pd.DataFrame(columns=['標題','時間','網址','內文']) #建立DataFrame
n = []
for b in pk:
    x=1
    while True:
        
        url = f'https://www.cardu.com.tw/news/list.php?nt_pk={b}&PageNo={x}'
        res = requests.get(url, headers = headers)
        soup = BeautifulSoup(res.text, 'html.parser')
        title= soup.select('h3')
        
        
        y=soup.select('div[class="col-12 page"] span')

        if y!=[]:
            if x == 1:
                max_page=int(re.findall(r'共([0-9]*)頁',str(y))[0]) #正規化找出頁數


            for i in title:

                article_title=i.a['title'] #標題
                print(article_title) 
                article_url=i.a['href'] #網址
                print(i.a['href'])


                url_a= i.a['href']
                res_a = requests.get(url_a, headers = headers)
                soup_a = BeautifulSoup(res_a.text, 'html.parser')
                title_a= soup_a.select('div[class="pb-3 mx-3 detail_content"]')
                posttime=soup_a.select('div[class="col-md-8 col-12"] p')


                a_text = ''.join(a.text for a in title_a) #內文
                print(a_text)        
                
                
                #timetext=''.join(p.text for p in posttime) #時間
                for p in posttime:
                    #print(p.text)
                    
                    pp=re.findall(r'\d\d\d\d-\d\d-\d\d',p.text)[0]
                    print(pp)
                    
                    

                text = ''
                for a in title_a:
                    text+= a.text


    #             for j in title_a:
    #                 print(j.text)
    #             print('-------------------------')

                article_title = rename(article_title)
                df.loc[index] = [article_title,pp,article_url,text] #把資料寫入DataFrame
                index += 1


#                 with open(r'%s/%s.txt' %(cardpath,article_title),'w',encoding='utf-8') as f:
#                     f.write(text)

            if (max_page == x) or (max_page == 0):
                break

            x += 1
        else:
            n.append(url)
```
