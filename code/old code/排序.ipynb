```python
top=pd.read_json(r'C:/Users/Big data/PycharmProjects/ourETL/top.json')
```

```python
card_l=[]
act_l=[]
for c in top['card']:
    for a in top['act']:
        if c not in card_l:
            card_l.append(c)
        if a not in act_l:
            act_l.append(a)
c_zero=[]
for i in range(len(card_l)):
    c_zero.append(0)
car_act=pd.DataFrame(columns=[card_l])
for i in act_l:
    car_act.loc[i]=c_zero
car_act=car_act.drop('',axis=0)
car_act=car_act.drop('',axis=1)

print(car_act)

for i in range(len(top)): 
    c=top['card'][i]
    a=top['act'][i]
    if a !='':
        if c !='':
            car_act.loc[a][c]+=1
        
print(car_act)
        
# table = pd.pivot_table(df, values='D', index=card_l,columns=act_l, aggfunc=np.sum, fill_value=0)
```
