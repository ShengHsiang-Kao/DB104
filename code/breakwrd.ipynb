```python
import pandas as pd
import os
import json
import jieba
import re
import xlsxwriter
from openpyxl import *
import operator
```

```python
jieba.load_userdict(r'C:/Users/Big data/PycharmProjects/ourETL/userdict.txt')
dfall=pd.read_json(r'C:/Users/Big data/PycharmProjects/ourETL/dfall.json')
```

```python
print(dfall)
```

```python
wb_list=[]
wc_list=[]
k=len(dfall)
for i in range(k):
    ja=dfall['內文'][i]
    gg=jieba.lcut(ja)
    wb_list.append(gg)
    word_c={}
    for i in gg:
        if i in word_c:
            word_c[i] +=1
        else:
            word_c[i] =1
    word_cnt=[(k,word_c[k])for k in word_c]
    word_cnt.sort(key=lambda x:x[1],reverse=True)
    wc_list.append(word_cnt)
# print(wb_list)
# print(wc_list)


dfall['斷詞']=wb_list
dfall['詞頻']=wc_list

# # print(df)
# # # # #
dfall.to_json('dfall.json',orient='index',force_ascii=False)

dfall.to_csv('dfall.csv',encoding='utf-8')
dfall.to_excel('dfall.xlsx')
# # # # # #


```
