```python

top=pd.DataFrame(columns=['bank','card','act'])

for num,b in enumerate(tag_df['bank']):
    if b=={}:
        top.loc[num]='','0','0'
    else:
        k,j=0,0
        for c_key in b:
            j=b[c_key]
            if j>k:
                k=j
                top.loc[num]=c_key,'0','0'
        
for num,b in enumerate(tag_df['card']):
    if b=={}:
        top.loc[num]=top['bank'][num],'','0'
    else:
        k,j=0,0
        for c in b:
            j=b[c]
            if j>k:
                k=j
                top.loc[num]=top['bank'][num],c,'0'

for num,b in enumerate(tag_df['act']):
    if b=={}:
        top.loc[num]=top['bank'][num],top['card'][num],''
    else:
        k,j=0,0
        for c in b:
            j=b[c]
            if j>k:
                k=j
                top.loc[num]=top['bank'][num],top['card'][num],c
```

```python
top.to_json(r'C:/Users/Big data/PycharmProjects/ourETL/top.json')
top.to_csv(r'C:/Users/Big data/PycharmProjects/ourETL/top.csv',encoding='utf-8')
top.to_excel(r'C:/Users/Big data/PycharmProjects/ourETL/top.xlsx')
```
