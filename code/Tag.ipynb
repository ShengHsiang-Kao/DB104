```python
jieba.load_userdict(r'C:/Users/Big data/PycharmProjects/ourETL/userdict.txt')
dfall=pd.read_json(r'C:/Users/Big data/PycharmProjects/ourETL/dfall.json')


for i in dfall['斷詞']:
    for n,j in enumerate(i):
        for k in bank:
            if j == k:
                i[n]=bank[k]
                
for i in dfall['斷詞']:
    for n,j in enumerate(i):
        for k in card:
            if j == k:
                i[n]=card[k]
for i in dfall['斷詞']:
    for n,j in enumerate(i):
        for k in behavior:
            if j == k:
                i[n]=behavior[k]
                
```

```python

# print(dfall)
# print(len(dfall['斷詞']))

bank_l=[]
for ii in range(len(dfall['斷詞'])):
    cont=dfall['斷詞'][ii]
#   print(ii)
    b_tag={}
    for i in cont:
        if i in bank:
            if i not in b_tag:
                b_tag[i]=1
            else:
                b_tag[i]+=1
    bank_l.append(b_tag)
print(bank_l)


card_l=[]
for ii in range(len(dfall['斷詞'])):
    cont=dfall['斷詞'][ii]
#   print(ii)
    c_tag={}
    for i in cont:
        if i in card:
            if i not in c_tag:
                c_tag[i]=1
            else:
                c_tag[i]+=1
    card_l.append(c_tag)
print(card_l)


act_l=[]
for ii in range(len(dfall['斷詞'])):
    cont=dfall['斷詞'][ii]
#   print(ii)
    act_tag={}
    for i in cont:
        if i in behavior:
            if i not in act_tag:
                act_tag[i]=1
            else:
                act_tag[i]+=1
    act_l.append(act_tag)
print(act_l)
```

```python
# print(len(act_l))
tag_df=pd.DataFrame(columns=['bank','card','act'])
for index in range(len(act_l)):
    tag_df.loc[index]=bank_l[index],card_l[index],act_l[index]
print(tag_df)

```
