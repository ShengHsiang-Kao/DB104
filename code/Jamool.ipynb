```python
age=[]
for n,i in enumerate(card_data.loc['年齡']):
#     print(i)   
    if i != '':
        if len(i)>30:
            a=re.findall(r'年齡.?.?.?(\d\d)',i)
            if a!=[]:
                print(a[0])
                age.append(a[0])
        else:
            num=re.findall(r'(\d\d)',i)
            if num !=[]:
                age.append(num[0])
            else:
                age.append('')
    else:
        age.append('')
# print(age)
# print(len(card_data.loc['年齡']))
# len(age)
# age[70]='35'

# age[111]='18'
age
len(age)
```

```python

service_t=[]
for n,i in enumerate(card_data.loc['年資']):
#     print(i)
    if i != '':
        mon=re.findall(r'\d.?.?.?.?月',i)
        if mon !=[]:
#             print(mon[0])
            service_t.append(mon[0])
            
        else:
            year=re.findall(r'\d.?.?.?.?年',i)
            if year!=[]:
#                 print(year[0])
                service_t.append(year[0])
            else:
#                 print(i,n)
                service_t.append('')
    
    else:
        service_t.append('')
service_t
```

```python
servT=[]
for m in service_t:
    m01=re.findall(r'(\d?.*\d?).*年(\d.*\d).*月',m)
    if m01!=[]:
        y_m=float(m01[0][0])*12+float(m01[0][1])
        print(y_m)
        servT.append(y_m)
    else:
        m02=re.findall(r'(\d?.*\d?).*年',m)
        if m02 != []:
            y=float(m02[0])*12
            print(y)
            servT.append(y)
        else:
            m03=re.findall(r'(\d?\d?).*月',m)
            if m03!=[]:
                mon=float(m03[0])
                print(mon)
                servT.append(mon)
            else:
                print(m)
                servT.append(m)

servT
```
