```python
dir_path='E:/專題/credit_card/DB104_hau/GPS/data'
results = [os.path.basename(f) for f in glob.glob(os.path.join(dir_path, '*.gpx'))]
all_gps_record=[]
for name in results:
    with open(r'./data/{}'.format(name),'r',encoding='utf-8') as gdata:
        gpx = gpxpy.parse(gdata)
    for track in gpx.tracks:
                for segment in track.segments:
                    for point in segment.points:
                        time = ""
                        if point.time:
                            time = point.time.isoformat()
                        gps_record={'user':name[:5],'time':time,'latitude':point.latitude,'longitude':point.longitude,'elevation':point.elevation,\
                                    'speed':point.speed,'source':point.source,'horizontal_dilution':point.horizontal_dilution,\
                                    'vertical_dilution':point.vertical_dilution,'course':point.course,'satellites':point.satellites}
                        all_gps_record.append(gps_record)
```

```python
col_name=['user','time','latitude','longitude','elevation','speed','source','horizontal_dilution','vertical_dilution','course','satellites']
user_track_df=pd.DataFrame(all_gps_record,columns=col_name)
user_track_df
```
