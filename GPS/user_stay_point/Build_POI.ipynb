```python
taipei= ["北投區","汐止區","林口區"]
```

```python
all_cinema_list=[]
for area in taipei:
    city_code=gmaps.geocode(area)
    city_info=city_code[0]['geometry']['location']
    cinema_list=gmaps.places_nearby(location=city_info,radius=10000,keyword='電影院',language='zh-TW')
    for result in cinema_list['results']:
        district=area
        name=result['name']
        latitude=result['geometry']['location']['lat']
        longtiude=result['geometry']['location']['lng']
        place_id=result['id']
        address=result['vicinity']
        cinema_info={'name':name,'district':'大台北'+district,'place_id':place_id,'address':address,'latitude':latitude,'longtiude':longtiude}
        all_cinema_list.append(cinema_info) 
```
