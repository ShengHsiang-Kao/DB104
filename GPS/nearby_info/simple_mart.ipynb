```python
simple_url = ('http://www.simplemart.com.tw/ec99/ushop20097/map.asp')
```

```python
driver = Chrome('./chromedriver')
driver.get(simple_url )

cities= ["台北市","新北市","桃園市","台中市","台南市","高雄市","基隆市","新竹市","嘉義市","新竹縣",
        "苗栗縣","彰化縣","南投縣","雲林縣","嘉義縣","屏東縣","宜蘭縣","花蓮縣","台東縣","澎湖縣"]
#driver.find_element_by_tag_name('html').send_keys(Keys.HOME)
time.sleep(1)
#//*[@id="pKey"]

#typeplace.
#//*[@id="pKey"]
#driver.find_element_by_xpath('/html/body/div[3]/ul/li[2]/a').click()
all_store=[]
sname=[]
saddr=[]
for i in cities:

    typeplace=driver.find_element_by_xpath('//*[@id="pKey"]')
    typeplace.clear()
    typeplace.send_keys(u'{}'.format(i))
    time.sleep(1)
    driver.find_element_by_id("btn-search").send_keys(Keys.RETURN)
    time.sleep(1)
    htmlsource = driver.page_source
    soup = BeautifulSoup(htmlsource, 'html.parser')
    name=soup.select('[class="map-name"]')
    addr=soup.select('[class="map-add"]')
    for n in name:
        sname.append(n.text)
    for a in addr:
        saddr.append(a.text)
    print(sname)
    print(saddr)
    
        
```
