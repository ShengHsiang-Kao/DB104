```python
import pandas as pd
from geopy.distance import geodesic
from datetime import datetime
import gmplot
#import time
```

```python
gps_data_df=pd.read_excel(r'./user_gps_data.xlsx',encoding='utf-16',index_col=0)
gps_data_df.head()
```

# 求兩點距離及速度

```python

#兩點距離求速度
localFormat = "%Y-%m-%d %H:%M:%S"
gps_data_df['point_speed']=0
gps_data_df['point_distance']=0
for user in list(set(gps_data_df['user'])):
    for i in gps_data_df[gps_data_df['user']==user].index:
        if i < gps_data_df[gps_data_df['user']==user].index[-1]:
            deltatime=datetime.strptime(gps_data_df.loc[i+1,'time'],localFormat)-datetime.strptime(gps_data_df.loc[i,'time'],localFormat)
            minutesdelta=(deltatime.seconds)/60
            if minutesdelta<10:
                prev=gps_data_df.loc[i,['latitude','longitude']]
                post=gps_data_df.loc[i+1,['latitude','longitude']]
                if minutesdelta>0:
                    gps_data_df.loc[i+1,'point_speed']=((geodesic(prev,post).kilometers)/minutesdelta)*60
                gps_data_df.loc[i+1,'point_distance']=(geodesic(prev,post).kilometers)
    #print(prev,post)
```

# 求staypoint 

```python
#門檻值
#3km/hr 15mins
#1.5km/hr 15mins
gps_data_df_complete=pd.read_excel(r'./gps_data_pointspeed.xlsx',encoding='utf-16',index_col=0)
gps_data_df_complete.head()
```

```python
speed_threshold=1.5
#if speed <1.5

#設定每個點時間，加總每個點時間
#total_time+=time_delta
#if total_time>10 minutes
#stay point.append(coord)
#if total time
#if speed>1.5 moving point


#計算移動距離



    

gps_data_df_complete['stay_point_time']=0
localFormat = "%Y-%m-%d %H:%M:%S"
for user in list(set(gps_data_df['user'])):
    stay_point=[]
    for i in gps_data_df[gps_data_df['user']==user].index:
        if i < gps_data_df[gps_data_df['user']==user].index[-1]:
            stay_time=0
            if gps_data_df_complete.loc[i,'point_speed']<speed_threshold:
                deltatime=datetime.strptime(gps_data_df.loc[i+1,'time'],localFormat)-datetime.strptime(gps_data_df.loc[i,'time'],localFormat)
                minutesdelta=(deltatime.seconds)/60
                stay_time +=minutesdelta
            stay_point.append(gps_data_df_complete[i,'latitude'],gps_data_df_complete[i,'longitude'])
                

            if minutesdelta<10:
                prev=gps_data_df.loc[i,['latitude','longitude']]
                post=gps_data_df.loc[i+1,['latitude','longitude']]
                if minutesdelta>0:
                    gps_data_df.loc[i+1,'point_speed']=((geodesic(prev,post).kilometers)/minutesdelta)*60
                gps_data_df.loc[i+1,'point_distance']=(geodesic(prev,post).kilometers)
```
