```python
def cut(random):
    # 切分訓練,測試
    global train_data, test_data, train_targets, test_targets,mean,std# 方便使用,設成全域
    train_data, test_data, train_targets, test_targets = train_test_split(df.loc[:, ["age","serveTime","Loan","SalPerY","holdCard","Career"]] , df.loc[:, ["credLimit_group"]] , test_size=0.3,random_state=random)
    # 標準化
    mean = train_data.mean(axis=0)
    train_data -=mean
    std = train_data.std(axis=0)
    train_data/=std
    test_data-=mean
    test_data/=std
    return train_data, test_data, train_targets, test_targets,mean,std
```
