# 關聯法則-文章card.behavior字詞有出現即算1次

```python
#準備data


#card
n=len(df1)
card_tag=[]
for article in range(n):
    article_card=set()
    for word in df1['seg'][article].split(' '):
            if word in card:
                article_card.add(card[word])
    card_tag.append(list(article_card))
card_tag[:10]               
#behavior
n=len(df1)
behavior_tag=[]
for article in range(n):
    article_behavior=set()
    for word in df1['seg'][article].split(' '):
            if word in behavior:
                article_behavior.add(behavior[word])
    behavior_tag.append(list(article_behavior))
                    
```

```python
alltag=[]
for i in range(50):
    each=card_tag[i]+behavior_tag[i]
    alltag.append(each)

    
```

```python
alltag[:10]
```

```python
te = TransactionEncoder()
te_ary = te.fit(alltag).transform(alltag)
df2 = pd.DataFrame(te_ary, columns=te.columns_)
frequent_itemsets = apriori(df2, min_support=0.3, use_colnames=True)

frequent_itemsets
```

```python
from mlxtend.frequent_patterns import association_rules

association_rules(frequent_itemsets, metric="confidence", min_threshold=0.7)
```

```python
from mlxtend.frequent_patterns import association_rules

association_rules(frequent_itemsets, metric="lift", min_threshold=2)
```

# 關聯法則-文章只計card.behavior出現最多之詞

```python
#準備data

#計算次數

#bank
n=len(df1)
bank_tag_wt=[]
#article_bank={}
for article in range(n):
    article_bank={}
    for word in df1['seg'][article].split(' '):
            if word in bank:
                if bank[word] not in article_bank:
                    article_bank[bank[word]]=1
                else:
                    article_bank[bank[word]]+=1
    #print(article_bank)
    bank_tag_wt.append(article_bank)
#card

card_tag_wt=[]
for article in range(n):
    article_card={}
    for word in df1['seg'][article].split(' '):
            if word in card:
                if card[word] not in article_card:
                    article_card[card[word]]=1
                else:
                    article_card[card[word]]+=1
    
    card_tag_wt.append(article_card)

#behavior

behavior_tag_wt=[]
for article in range(n):
    article_behavior={}
    for word in df1['seg'][article].split(' '):
            if word in behavior:
                if behavior[word] not in article_behavior:
                    article_behavior[behavior[word]]=1
                else:
                    article_behavior[behavior[word]]+=1
    behavior_tag_wt.append(article_behavior)
                    
```

```python
max_list=[]
e={}
for i in range(len(card_tag_wt)):
    article_item=[]
    if card_tag_wt[i] ==e:
        pass
    else:
        article_item.append(max(card_tag_wt[i].items(), key=lambda a:a[1],default=0)[0])
    if behavior_tag_wt[i] ==e:
        pass
    else:
        article_item.append(max(behavior_tag_wt[i].items(), key=operator.itemgetter(1),default=0)[0])
    max_list.append(article_item)
```

```python
#關聯法則
te2 = TransactionEncoder()
te_ary2 = te2.fit(max_list).transform(max_list)
df3 = pd.DataFrame(te_ary2, columns=te2.columns_)
frequent_itemsets2 = apriori(df3, min_support=0.0001, use_colnames=True)

frequent_itemsets2
```

```python
from mlxtend.frequent_patterns import association_rules

association_rules(frequent_itemsets2, metric="confidence", min_threshold=0.1)
```

```python
from mlxtend.frequent_patterns import association_rules

association_rules(frequent_itemsets2, metric="lift", min_threshold=3)
```
