```python
import jieba
import json
import pandas as pd
import os
```

```python
df=pd.read_excel(r'D:/Google雲端硬碟/資料分析師班/專題/credit_card/credit_card/youtube_crawl/text_file/video_df2ex_utf16.xlsx',encoding='utf-16')
```

```python
df1=df
df1['seg']=None
```

```python
df1['word_freq']=None
```

```python
n=len(df)
alltext=[]
stopset=set()
stop2=['/n','']


with open('C:/Users/user/Desktop/project_test/youtube_crawl/text_file/stop.txt','r',encoding='utf-8') as s:
    for line in s:
        stopset.add(line.strip('\n'))
jieba.load_userdict('userdict.txt')
for i in range(n):
    seg=''
    wf={}
    text= str(df1.loc[i]['title'])+'\n'+str(df1.loc[i]['description'])+'\n'+str(df1.loc[i]['comment'])+'\n'
    cut=jieba.cut(text,cut_all=False)
    for j in cut:
        if j not in stopset:
            seg +=j+' '
    df1['seg'][i]=seg
    for w in seg.split(' '):
        if w not in wf:
            wf[w]=1
        else:
            wf[w]+=1
    word_list=[(k,wf[k]) for k in wf if k not in stop2 ]
    word_list.sort(key=lambda a :a[1],reverse=True)
    df1['word_freq'][i]=word_list

    
```

```python
df1.to_excel(r'./video_df2ex_utf16_seg.xlsx','w',encoding='utf-16')
```
