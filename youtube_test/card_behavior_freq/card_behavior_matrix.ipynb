# 出現bank,card,behavior字詞計算1次

```python
#bank
n=len(df1)
bank_tag=[]
for article in range(n):
    article_bank=set()
    for word in df1['seg'][article].split(' '):
            if word in bank:
                article_bank.add(bank[word])
    bank_tag.append(list(article_bank))
print(bank_tag[:10])
                    
```

```python
#card
n=len(df1)
card_tag=[]
for article in range(n):
    article_card=set()
    for word in df1['seg'][article].split(' '):
            if word in card:
                article_card.add(card[word])
    card_tag.append(list(article_card))
card_tag[:10]               
```

```python
#behavior
n=len(df1)
behavior_tag=[]
for article in range(n):
    article_behavior=set()
    for word in df1['seg'][article].split(' '):
            if word in behavior:
                article_behavior.add(behavior[word])
    behavior_tag.append(list(article_behavior))
                    
```

```python
#合併dataframe

d={'bank_tag':bank_tag,'card_tag':card_tag,'behavior_tag':behavior_tag}
df_tag=pd.DataFrame(data=d)
df_tag
```

```python
#計算card behavior出現次數矩陣
all_bank=set([bank[b] for b in bank])
all_card=set([card[c] for c in card])
all_behavior=set([behavior[b] for b in behavior])


#建立df並填0
df_card_behavoir=pd.DataFrame(index=all_card,columns=all_behavior)
for i in df_card_behavoir.columns:
    df_card_behavoir[i]=0
df_card_behavoir
```

```python
#計算
ntag=len(df_tag)
#填入df之value
for i in  range(ntag):
    for c_tag in df_tag['card_tag'][i]:
        for b_tag in df_tag['behavior_tag'][i]:
            df_card_behavoir[b_tag].loc[c_tag]+=1
        
```

# 增加出現字詞次數為權重

```python
#計算次數

#bank
n=len(df1)
bank_tag_wt=[]
#article_bank={}
for article in range(n):
    article_bank={}
    for word in df1['seg'][article].split(' '):
            if word in bank:
                if bank[word] not in article_bank:
                    article_bank[bank[word]]=1
                else:
                    article_bank[bank[word]]+=1
    #print(article_bank)
    bank_tag_wt.append(article_bank)
#card

card_tag_wt=[]
for article in range(n):
    article_card={}
    for word in df1['seg'][article].split(' '):
            if word in card:
                if card[word] not in article_card:
                    article_card[card[word]]=1
                else:
                    article_card[card[word]]+=1
    
    card_tag_wt.append(article_card)

#behavior

behavior_tag_wt=[]
for article in range(n):
    article_behavior={}
    for word in df1['seg'][article].split(' '):
            if word in behavior:
                if behavior[word] not in article_behavior:
                    article_behavior[behavior[word]]=1
                else:
                    article_behavior[behavior[word]]+=1
    behavior_tag_wt.append(article_behavior)
                    
```

```python
#合併成df
data_wt={'bank_tag_wt':bank_tag_wt,'card_tag_wt':card_tag_wt,'behavior_tag_wt':behavior_tag_wt}
df_tag_wt=pd.DataFrame(data=data_wt)
type(df_tag_wt['bank_tag_wt'].loc[1])
```

```python
#建立df_wt並放0
df_card_behavoir_wt=pd.DataFrame(index=all_card,columns=all_behavior)
for i in df_card_behavoir_wt.columns:
    df_card_behavoir_wt[i]=0
df_card_behavoir_wt
```

```python
#計算各列 card及behavior出現次數
ntag=len(df_tag_wt)
card_sum_word=[]
behavior_sum_word=[]
for i in  range(ntag):
    card_word=0
    for c in df_tag_wt['card_tag_wt'][i]:
        card_word+=df_tag_wt['card_tag_wt'][i][c]
    card_sum_word.append(card_word)
    behavior_word=0
    for c in df_tag_wt['behavior_tag_wt'][i]:
        behavior_word+=df_tag_wt['behavior_tag_wt'][i][c]
    behavior_sum_word.append(behavior_word)
print(card_sum_word[:20])
print(behavior_sum_word[:20])

#加入各文章字詞總數於df
df_tag_wt['card_sum_word']=card_sum_word
df_tag_wt['behavior_sum_word']=behavior_sum_word
df_tag_wt['total_freq']=0
for f in range(len(df_tag_wt)):
    df_tag_wt['total_freq'][f]=df_tag_wt['card_sum_word'][f]+df_tag_wt['behavior_sum_word'][f]
```

```python
#填入df之value加上權重
ntag=len(df_tag_wt)

for i in  range(ntag):
    nword=0
    for c_tag in df_tag_wt['card_tag_wt'][i]:
        for b_tag in df_tag_wt['behavior_tag_wt'][i]:
            df_card_behavoir_wt[b_tag].loc[c_tag] += ((df_tag_wt['card_tag_wt'][i][c_tag]/df_tag_wt['total_freq'][i])+(df_tag_wt['behavior_tag_wt'][i][b_tag]/df_tag_wt['total_freq'][i]))
```

```python
#round
for  i in df_card_behavoir_wt.columns:
    df_card_behavoir_wt[i]=round(df_card_behavoir_wt[i],2)
```
