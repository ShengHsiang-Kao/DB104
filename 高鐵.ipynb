```python
import pandas as pd
from datetime import datetime
from geopy.distance import geodesic
import numpy as np
```

```python
gps_data_df=pd.read_excel('/Users/roland/user_gps_data.xlsx')

```

```python
#算出位置點及速度點
localFormat = "%Y-%m-%d %H:%M:%S"
gps_data_df['point_speed']=0
gps_data_df['point_distance']=0
for user in list(set(gps_data_df['user'])):
    for i in gps_data_df[gps_data_df['user']==user].index:  #每個user的index
        if i < gps_data_df[gps_data_df['user']==user].index[-1]:
            deltatime=datetime.strptime(gps_data_df.loc[i+1,'time'],localFormat)-datetime.strptime(gps_data_df.loc[i,'time'],localFormat)
            minutesdelta=(deltatime.seconds)/60  #上下兩個點時間差
            if minutesdelta<10:
                prev=gps_data_df.loc[i,['latitude','longitude']]
                post=gps_data_df.loc[i+1,['latitude','longitude']]
                if minutesdelta>0:
                    gps_data_df.loc[i+1,'point_speed']=((geodesic(prev,post).kilometers)/minutesdelta)*60
                gps_data_df.loc[i+1,'point_distance']=(geodesic(prev,post).kilometers)
    print(prev,post)
```

```python
#把每個user速度包成list 再把所有user包成大list
user=sorted(list(set(gps_data_df['user'])))#把user做成list
allspeed=[]
for j in user:
    Uspeed=[]
    for i in range(len(gps_data_df)):
        if gps_data_df.loc[i]['user']==j:
            k=(j,gps_data_df.loc[i]['point_speed'])
            Uspeed.append(k)#每行帶進來如果是同user就append
    allspeed.append(Uspeed)
allspeed
```

```python
#把同user的資料騎車的值挑出
trafficnum=[]
for i in allspeed:
    count=0
    for ii in i:
        if ii[1]>15:   #以時速>15為騎車回傳值每次＋1
            count+=1
    uct=(ii[0],count)
    trafficnum.append(uct)
trafficnum
```

```python
#如果開車時間大於30的話推薦加油
gas_recmd=[]
for i in trafficnum:
    if i[1]>30:
        gas_recmd.append(i[0])
gas_recmd
```

```python
#計算超過時速超過150的次數 推薦高鐵
hspeed=[]
for i in allspeed:
    count=0
    for ii in i:
        if ii[1]>150:
            count+=1
    uct=(ii[0],count)
    hspeed.append(uct)
print(hspeed)
Hw_recmd=[]
for i in hspeed:
    if i[1]>0:
        Hw_recmd.append(i[0])
Hw_recmd
```

```python
#以天數和user做包含經緯度的list
travel=[]
trv_user=[]
tra_all=[]
k1=0
user1=0
for n,i in enumerate(gps_data_df['time']): #取日期為Ｋ
    k=i.split(' ')[0]
    user=gps_data_df['user'][n]
    lat=gps_data_df['latitude'][n]
    log=gps_data_df['longitude'][n]
    if (k1==k and user1==user):  #如果Ｋ相同放到就加在同一個list(travel)資料為人物,日期,經緯度
        travel.append((user,k,lat,log))        
    else:         #不同的排除第一次以外把travel加到大list(tra_all)
        if travel!=[]:
            tra_all.append(travel)
        travel=[]
        if user!=user1:  #如果user不同把大tra_all加進trv_user
            if tra_all!=[]:
                trv_user.append(tra_all)
                tra_all=[]
        travel.append((user,k,lat,log)) #第一次遇到不同的時候就可以開始放travel
        k1=k           #把k1變為這次的日期
        user1=user
tra_all.append(travel)#最後一次放完travel在這裏加進大list
trv_user.append(tra_all)
trv_user
```

```python
#計算出遊距離並找出人物及天數
lat=[];lng=[];traveldata=[];travDis_all=[];medall=[]
for n,i in enumerate(trv_user): #每個user為一筆帶進去
    travDis=[]
    diday=[]
    for n2,i2 in enumerate(i): #每一天為一筆帶進去
        lat=[]
        lng=[]
        
        for n3,i3 in enumerate(i2): #每個數據為i3 把所有的lat做個list 緯度too
            lat.append(i3[2])
            lng.append(i3[3])
        a=lat.index(max(lat))
        b=lat.index(min(lat))
        latMax=lat[a];latMin=lat[b] #取lat的最大經緯度 和lat 最小的經緯度
        a1=lng.index(max(lng))
        b1=lng.index(min(lng))
        lngMax=lat[a1];lngMin=lat[b1] #取lng的最大經緯度 和lng 最小的經緯度
        
        #最大移動經度和最大移動緯度取最大值
        distance = max(geodesic(latMax,latMin).kilometers,geodesic(lngMax,lngMin).kilometers) 
        diday.append(distance)#距離list計算用
        distance=(distance,i2[n3][1],i2[n3][0])#資料list查找用
        travDis.append(distance)#所有資料以天為主匯入
    travDis_all.append(travDis)#所有資料以人為主匯入
    medtravel=np.median(diday)#計算中位數移動距離
    medall.append(medtravel)#每個人的距離的中位數
for n4,i4 in enumerate(travDis_all): #每人資料i4
    CT=0
    for i5 in i4: #每日資料i5
        if i5[0]>medall[n4]+3: #超過中位數三公里算出遊CT每次＋1
            CT+=1
    print('人物:',i5[2],'出遊天數',CT)    
    
```
