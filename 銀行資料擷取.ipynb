```python
#檢查那些銀行沒有在大表中並做INDEX表
allind=[]
for i in data01['銀行']:
    index=[]
    for n,i2 in enumerate(data02['銀行']):
        if i == i2:
            index.append(n) #銀行名字相同打卡的INDEX放在同個LIST
    
    if index==[]:
        print(i)
    allind.append(index)#小表卡的銀行和大表卡的INDEX呈現
```

```python
#出現在小表卡的字 有幾個字出現在大表該銀行的卡中
allmax=[]
for n,icarstr in enumerate(data01['卡名']):
    icarstr=str(icarstr)
    gofor=[]
    #在大表該銀行的卡中所有卡一張一張去比對小表的字
    for n2,icarname in enumerate(allind[n]):
        num=0
        #小表該卡每個字出現+1並放進list
        for st in icarstr:
            if st in data02['卡片名稱'][icarname]:
                num+=1
        gofor.append(num)
        
    allmax.append(gofor)
#         aaaa=gofor.index(max(gofor))
print(allmax)
#去出所有字出在該卡(大表中)現的次數

```

```python
#做個list把所有出現數字最高的卡放進來max!
jjj=[]
for n,i in enumerate(allmax):
    if i!=[]:
        bb=max(i)
        
        bb=allmax[n].index(bb)
    else:
        bb=''
    jjj.append(bb)
```
