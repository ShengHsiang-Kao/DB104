# 資料清理

```python
#把全形字或標點符號轉成半形, 轉成半形之後的標點符號才可被isalpha()清掉
def conv_wide(text):
    code = ord(str(text))
    if x == u'\u3000': #space
        code = 32
    elif u'\uff01' <= text <= u'\uff5e':
        code -= 65248
    return chr(code)
```

```python
#articles去除所有的標點符號
from datetime import datetime
for k in range(len(articles)):
    temp = []
    print(k)
    try:
        for wd in str(articles[k]):
            x = wd.upper()
            y = conv_wide(x)
            if x.isalpha():
                temp.append(y)
            new_article = ''.join(temp)   
        articles[k] = new_article
    except Exception as e:
        print (e.__doc__) 
```

# 斷詞

```python
import jieba
jieba.load_userdict(r'./userdict.txt')
for j in range(len(articles)):
    try:
        cut[j] = jieba.lcut(str(articles[j]), cut_all=False) #jieba.lcut 直接返回 list, 而不是 generator
    except Exception as e:
        print (e.__doc__)
    cut_list = list()
    for wd in cut[j]:
        if wd != ' ':
            cut_list.append(wd)
```

```python
from gensim.models import FastText
model = FastText(size=100, window=20, min_count=1)
model.build_vocab(sentences=cut)
model.train(sentences=cut, total_examples=len(cut), epochs=10)
```
