```python
dataset = [['Milk', 'Onion', 'Nutmeg', 'Kidney Beans', 'Eggs', 'Yogurt'],
           ['Dill', 'Onion', 'Nutmeg', 'Kidney Beans', 'Eggs', 'Yogurt'],
           ['Milk', 'Apple', 'Kidney Beans', 'Eggs'],
           ['Milk', 'Unicorn', 'Corn', 'Kidney Beans', 'Yogurt'],
           ['Corn', 'Onion', 'Onion', 'Kidney Beans', 'Ice cream', 'Eggs']]

te = TransactionEncoder()
te_ary = te.fit(dataset).transform(dataset)
df = pd.DataFrame(te_ary, columns=te.columns_)
frequent_itemsets = apriori(df, min_support=0.6, use_colnames=True)

frequent_itemsets
```

```python
#bank
n=len(df1)
bank_tag_wt=[]
#article_bank={}
for article in range(n):
    article_bank={}
    for word in df1['seg'][article].split(' '):
            if word in bank:
                if bank[word] not in article_bank:
                    article_bank[bank[word]]=1
                else:
                    article_bank[bank[word]]+=1
    #print(article_bank)
    bank_tag_wt.append(article_bank)
#card

card_tag_wt=[]
for article in range(n):
    article_card={}
    for word in df1['seg'][article].split(' '):
            if word in card:
                if card[word] not in article_card:
                    article_card[card[word]]=1
                else:
                    article_card[card[word]]+=1
    
    card_tag_wt.append(article_card)

#behavior

behavior_tag_wt=[]
for article in range(n):
    article_behavior={}
    for word in df1['seg'][article].split(' '):
            if word in behavior:
                if behavior[word] not in article_behavior:
                    article_behavior[behavior[word]]=1
                else:
                    article_behavior[behavior[word]]+=1
    behavior_tag_wt.append(article_behavior)
                    
```

```python
data_wt={'bank_tag_wt':bank_tag_wt,'card_tag_wt':card_tag_wt,'behavior_tag_wt':behavior_tag_wt}
df_tag_wt=pd.DataFrame(data=data_wt)
type(df_tag_wt['bank_tag_wt'].loc[1])
```

```python
#建立df_wt
df_card_behavoir_wt=pd.DataFrame(index=all_card,columns=all_behavior)
```

```python
#df_wt放入0
for i in df_card_behavoir_wt.columns:
    df_card_behavoir_wt[i]=0
df_card_behavoir_wt
```

```python
#計算各列 card及behavior出現次數
ntag=len(df_tag_wt)
card_sum_word=[]
behavior_sum_word=[]
for i in  range(ntag):
    card_word=0
    for c in df_tag_wt['card_tag_wt'][i]:
        card_word+=df_tag_wt['card_tag_wt'][i][c]
    card_sum_word.append(card_word)
    behavior_word=0
    for c in df_tag_wt['behavior_tag_wt'][i]:
        behavior_word+=df_tag_wt['behavior_tag_wt'][i][c]
    behavior_sum_word.append(behavior_word)
print(card_sum_word[:20])
print(behavior_sum_word[:20])
```

```python
#加入各文章字詞總數於df
df_tag_wt['card_sum_word']=card_sum_word
df_tag_wt['behavior_sum_word']=behavior_sum_word
df_tag_wt['total_freq']=0
```

```python
for f in range(len(df_tag_wt)):
    df_tag_wt['total_freq'][f]=df_tag_wt['card_sum_word'][f]+df_tag_wt['behavior_sum_word'][f]
```
