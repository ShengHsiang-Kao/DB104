```python
import pandas as pd
import os
import jieba
source_file_path = r'./res_creditcard'
file_list = os.listdir(source_file_path)
file_list
```

```python
df = pd.DataFrame(columns = ['編號','標題','時間','網址','內文'])
df
```

```python
for n, each_article in enumerate(file_list):
    article_path = source_file_path + '/' + each_article
    with open(article_path, 'r', encoding = 'utf-8') as f:
        tmp_article_string = f.read()
    try:
        tmp_row_data = tmp_article_string.split('看板creditcard標題')[1].split('---split---')
        cont=tmp_row_data[0]
        info_1=tmp_row_data[1]
        k=info_1.split('\n')
        title=k[5].split(':')[1]
        time=k[6].split(':')[1]
        link=k[7].split('網址:')[1]
        serial=n+1
#         print(title)
#         print(time)
#         print(link)
    except:
        print(n)
    df.loc[n+1]= serial,title,time,link,cont
#     print(n)
```

```python
# 以'核卡'和'心得'作為關鍵字，找出符合核卡心得的文章編號
import jieba
import pandas as pd
import re
df=pd.read_json (r'./ptt_creditcard.json')
jieba.load_userdict('./userdict.txt')
# print(df)
title=df.loc['標題']
ti_ind=[]
for idx,i in enumerate(title):
    a=re.findall(r'.*核卡.*',i)
    if a!=[]:
        ti_ind.append(idx+1)
    a=re.findall(r'.*心得.*',i)
    if a!=[]:
        if idx+1 not in ti_ind:
            ti_ind.append(idx+1)
print(ti_ind)
# for z in ti_ind:
#     print(df.loc['標題'][z])
```

```python
import jieba
jieba.load_userdict('./userdict.txt')
gg_list=[]
word_clist=[]
content=df.loc['內文']
for j in range(1,len(content)+1):
    gg=jieba.lcut(content[j])
#     print(gg)
    gg_list.append(gg)
    
    word_c={}
    for i in gg:
        if i in word_c:
            word_c[i] +=1
        else:
            word_c[i] =1
    word_clist.append(word_c)
# print('gg_list')
print(len(gg_list))
print(len(word_clist))
print(len(content))
df.loc['斷詞']=gg_list
df.loc['詞頻']=word_clist
```

```python
df
df['銀行']=None
df['卡片']=None
df['行為']=None
```

```python
for article in range(1,len(df)+1):
    article_bank={}
    for word in df['斷詞'][article]:
        if word in bank:
            if bank[word] not in article_bank:
                article_bank[bank[word]] = 1
            else:
                article_bank[bank[word]] += 1
    print(article_bank)
    df['銀行'][article]=article_bank   
```

```python
for article in range(1,len(df)+1):
    article_card={}
    for word in df['斷詞'][article]:
        if word in card:
            if card[word] not in article_card:
                article_card[card[word]] = 1
            else:
                article_card[card[word]] += 1
    print(article_card)
    df['卡片'][article]=article_card
```

```python
for article in range(1,len(df)+1):
    article_behavior={}
    for word in df['斷詞'][article]:
        if word in behavior:
            if behavior[word] not in article_behavior:
                article_behavior[behavior[word]] = 1
            else:
                article_behavior[behavior[word]] += 1
    print(article_behavior)
    df['行為'][article]=article_behavior
```
