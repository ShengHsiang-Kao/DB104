```python
from gensim.models import FastText
import pandas as pd
import json
df_corpus1 = pd.read_json('E:/my/group_project/ptt_clean_cut_articles.json')
df_corpus2 = pd.read_json('E:/my/group_project/dcard_clean_cut_articles.json')
df_corpus3 = df_corpus2.drop(['詞頻'], axis=0)
df_corpus = pd.concat([df_corpus1, df_corpus3], ignore_index=True, axis=1)
len(df_corpus.columns)
```

```python
df_corpus
```

```python
articles = df_corpus.loc['內文']
df_corpus.loc['斷詞']='0'
cut = df_corpus.loc['斷詞']
print(cut)
```

```python
import jieba
jieba.load_userdict(r'./userdict.txt')
for j in range(len(articles)):
    try:
        cut[j] = jieba.lcut(str(articles[j]), cut_all=False) #jieba.lcut 直接返回 list, 而不是 generator
    except Exception as e:
        print (e.__doc__)
    cut_list = list()
    for wd in cut[j]:
        if wd != ' ':
            cut_list.append(wd)
```

```python
cut
```

```python
model = FastText(size=100, window=20, min_count=1)
model.build_vocab(sentences=cut)
model.train(sentences=cut, total_examples=len(cut), epochs=10)
```

```python
model.wv.most_similar('機票')
```
